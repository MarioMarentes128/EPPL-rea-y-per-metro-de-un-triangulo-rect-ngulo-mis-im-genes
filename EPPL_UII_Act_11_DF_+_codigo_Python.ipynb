```python
# Calcular el area de un triangulo rectangulo
# Variables h, b, A
# Formulas A = b * h / 2
# Capturar el valor de A

print("Calcular Area y perimetro de un triangulo rectangulo")
h = float(input("Ingrese el valor de la altura: "))
b = float(input("Ingrese el valor de la base: "))
A = b * h / 2
print("El area del triangulo rectangulo es: ", A)
print("Mario Yañez NL = 53")
```
